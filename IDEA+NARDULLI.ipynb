#### Importe de bibliotecas esenciales

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

#### Carga del dataset a evaluar

```python
data = pd.read_csv('Data/Ames_Housing_Data.csv')
data.head()
```

```python
data.shape
```

#### Busqueda de correlaciones 

```python
plt.figure(dpi = 120,figsize= (25,25))
mask = np.triu(np.ones_like(data.corr(),dtype = bool))
sns.heatmap(data.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
plt.yticks(rotation = 0)
plt.xticks(rotation = 90)
plt.title('Correlation Heatmap')
plt.show()
```

#### antes de realizar el join plot dropearemos todas las columnas que no son numerica de nuestro dataset 

```python
data = data.select_dtypes(['number'])
```

```python
data.shape
```

#### realizamos otra busqueda de correlaciones

```python
plt.figure(dpi = 120,figsize= (25,25))
mask = np.triu(np.ones_like(data.corr(),dtype = bool))
sns.heatmap(data.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
plt.yticks(rotation = 0)
plt.xticks(rotation = 90)
plt.title('Correlation Heatmap')
plt.show()
```

#### Join Plot

```python
plt.figure(dpi = 100, figsize = (5,4))
print("Joint plot de Precio de venta con otras variables ==> \n")
biggerCorr = -100
correlacion = ''
for i in  data.columns:
    if i != 'SalePrice':
        print(f"Correlacion entre Precio de venta y {i} ==> ", data.corr().loc['SalePrice'][i])
        sns.jointplot(x='SalePrice', y=i, data=data, kind = 'reg', color = 'purple')
        plt.show()
        if (biggerCorr < data.corr().loc['SalePrice'][i]):
            biggerCorr = data.corr().loc['SalePrice'][i]
            correlacion = f"Correlacion entre Precio de venta y {i} ==> " + str(data.corr().loc['SalePrice'][i])
print('La mayor correlacion es la ' + correlacion)
```
